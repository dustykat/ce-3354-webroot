## Exercise 2:
Use the results in Exercise 1 (above) to prepare Intensity-Duration-Frequency curves for Eden, Texas (Concho County). The desired durations are 30-min, 1-hour, 2-hour, 3-hour, 6-hour, 12-hour, and 24-hour. The desired ARI are 2-yr, 10-yr, 50-yr, and 100-yr. You will produce a total of 4 curves.

Solution:

```python
# 2-yr

duration = [5/60,
10/60,
15/60,
30/60,
60/60,
2,
3,
6,
12,
24
]
depth = [0.412,
0.662,
0.82,
1.13,
1.45,
1.76,
1.94,
2.26,
2.63,
3.04
]
intensity = []
for i in range(len(depth)):
    intensity.append(depth[i]/duration[i])
ari = "2-year"
from matplotlib import pyplot as plt # import the plotting library from matplotlibplt.show()
plt.plot( duration, intensity, color ='green', marker ='o', linestyle ='solid') # create a line chart, years on x-axis, gdp on y-axi
plt.yscale('log')
plt.xscale('log')
plt.title("DDF near Eden, Texas ARI = " + ari)# add a title
plt.xlabel("Duration (hours)")# add a label to the x and y-axes
plt.ylabel("Intensity (inches/hour)")
plt.show() # display the plot

# 10-yr

duration = [5/60,
10/60,
15/60,
30/60,
60/60,
2,
3,
6,
12,
24
]
depth = [0.643,
1.03,
1.28,
1.75,
2.27,
2.82,
3.15,
3.76,
4.4,
5.09
]
intensity = []
for i in range(len(depth)):
    intensity.append(depth[i]/duration[i])
ari = "10-year"
from matplotlib import pyplot as plt # import the plotting library from matplotlibplt.show()
plt.plot( duration, intensity, color ='blue', marker ='o', linestyle ='solid') # create a line chart, years on x-axis, gdp on y-axi
plt.yscale('log')
plt.xscale('log')
plt.title("DDF near Eden, Texas ARI = " + ari)# add a title
plt.xlabel("Duration (hours)")# add a label to the x and y-axes
plt.ylabel("Intensity (inches/hour)")
plt.show() # display the plot

# 50-yr

duration = [5/60,
10/60,
15/60,
30/60,
60/60,
2,
3,
6,
12,
24
]
depth = [0.86,
1.39,
1.7,
2.32,
3.03,
3.87,
4.41,
5.38,
6.39,
7.43,
]
intensity = []
for i in range(len(depth)):
    intensity.append(depth[i]/duration[i])
ari = "50-year"
from matplotlib import pyplot as plt # import the plotting library from matplotlibplt.show()
plt.plot( duration, intensity, color ='orange', marker ='o', linestyle ='solid') # create a line chart, years on x-axis, gdp on y-axi
plt.yscale('log')
plt.xscale('log')
plt.title("DDF near Eden, Texas ARI = " + ari)# add a title
plt.xlabel("Duration (hours)")# add a label to the x and y-axes
plt.ylabel("Intensity (inches/hour)")
plt.show() # display the plot

# 100-yr

duration = [5/60,
10/60,
15/60,
30/60,
60/60,
2,
3,
6,
12,
24
]
depth = [0.949,
1.53,
1.88,
2.55,
3.35,
4.33,
4.99,
6.15,
7.35,
8.58,
]
intensity = []
for i in range(len(depth)):
    intensity.append(depth[i]/duration[i])
ari = "100-year"
from matplotlib import pyplot as plt # import the plotting library from matplotlibplt.show()
plt.plot( duration, intensity, color ='red', marker ='o', linestyle ='solid') # create a line chart, years on x-axis, gdp on y-axi
plt.yscale('log')
plt.xscale('log')
plt.title("DDF near Eden, Texas ARI = " + ari)# add a title
plt.xlabel("Duration (hours)")# add a label to the x and y-axes
plt.ylabel("Intensity (inches/hour)")
plt.show() # display the plot
```

## Exercise 3:

Solution: 

Use script from lecture , adjust to correct depth for Eden, Texas

```python
# SCS design storm

hour = [0,2,4,6,7,8,8.5,9,9.5,9.75,10,10.5,11,11.5,11.75,12,12.5,13.0,13.6,14,16,20,24]
type2 = [0,0.022,0.048,0.08,0.098,0.12,0.133,0.147,0.163,0.172,0.181,0.204,0.235,0.283,0.357,0.663,0.735,0.772,0.799,0.82,0.88,0.952,1]

Ptotal = 7.43
T2D =[0 for i in range(len(hour))]
for i in range(len(hour)):
    T2D[i]=Ptotal*type2[i]
#plot

import matplotlib.pyplot as plt # the python plotting library
plottitle ='SCS Type 2 Design Storm for Eden, Texas for Total Depth =' + str(Ptotal) +' inches'
mydata = plt.figure(figsize = (10,5)) # build a square drawing canvass from figure class
plt.plot(hour, T2D, c='orange') # step plot
#plt.plot(time, accumulate, c='blue',drawstyle='steps') # step plot
plt.xlabel('Time (hours)')
plt.ylabel('Depth (inches)')
plt.title(plottitle)
plt.show()
```
